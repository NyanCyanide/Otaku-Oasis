```python
# preprocessing the text
def preprocess_text(text):
    # Tokenize the text
    text = str(text)
    tokens = word_tokenize(text.lower())

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Stem the words
    stemmer = SnowballStemmer('english')
    tokens = [stemmer.stem(token) for token in tokens]

    # Join the tokens back into a single string
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

# Pre-process the title and synopsis columns
anime_data['title'] = anime_data['title'].apply(preprocess_text)
anime_data['synopsis'] = anime_data['synopsis'].apply(preprocess_text)
print("Preprocessing done")
```

```python
def build_inverted_index(documents):
    inverted_index = defaultdict(list)

    for doc_id, document in enumerate(documents):
        terms = document.split()
        for term in set(terms):
            inverted_index[term].append(doc_id)

    return inverted_index


# Build the inverted index for the title and synopsis columns
title_inverted_index = build_inverted_index(anime_data['title'])
synopsis_inverted_index = build_inverted_index(anime_data['synopsis'])
print("Inverted index done")
```

```python
# Boolean model
def boolean_search(query, inverted_index):
    # Split the query into terms
    terms = query.split()

    # Find the documents that contain each term
    documents = [set(inverted_index[term]) for term in terms]

    # Combine the documents using set operations
    result = set()
    if len(documents) > 0:
        result = documents[0]
        for i in range(1, len(documents)):
            if len(documents[i]) > 0:
                if terms[i - 1] == 'and':
                    result = result.intersection(documents[i])
                elif terms[i - 1] == 'or':
                    result = result.union(documents[i])
                elif terms[i - 1] == 'not':
                    result = result.difference(documents[i])

    # Convert the result to a list and return it
    return list(result)
```
